### Apartado 2-a: Mostrar la cantidad y la media de películas producidas por año x plataforma. Crear un gráfico que muestre los resultados por plataforma.

```python
# Cargamos el archivo y lo abrimos con un Dataframe:

PATH2 = 'Datos/MoviesOnStreamingPlatforms_updated.csv'
df_parte2 = pd.read_csv(PATH2)
```

```python
df_parte2
```

```python
# vamos a obtener la información del dataframe para ver si tenemos datos nulos que nos puedan afectar y para ver los 
# tipos de datos.

# El primer problema que nos excontramos es que hay columnas con datos nulos (Age, IMBd, Rotten, Directors, 
# Genres, Cuntries, Language y Runtime)

df_parte2.info()
```

```python
df_parte2.describe()
```

```python
# Filtramos por las columnas de las plataformas, las agrupamos por año y contamos las películas por plataforma


df_parte2_plataformas = df_parte2.loc[:, ["Year", "Netflix", "Hulu", "Prime Video", "Disney+"]]
df_parte2_plataformas = df_parte2_plataformas.groupby(by="Year").sum()
df_parte2_plataformas
```

```python
# Una vez obtenido el Daframe que nos interesa, hacemos el gráfico:

df_parte2_plataformas.plot()

plt.title("Películas por año y plataforma unificado", size = 10)
```

```python
# Como se observa, las gráficas por plataforma coinciden en gran parte.
# Vamos a crear un nuevo gráfico que muestre el resultado por plataforma de forma independiente:


df_parte2_plataformas.plot.bar(rot = 1, subplots = True, fontsize = 7, figsize = (20, 10))
```

```python
# Hacemos lo mismo con las medias:

df_parte2_plataformas_medias = df_parte2.loc[:, ["Year", "Netflix", "Hulu", "Prime Video", "Disney+"]]
df_parte2_plataformas_medias = df_parte2_plataformas_medias.groupby(by="Year").mean()
df_parte2_plataformas_medias
```

```python
# Hacemos el gráfico correspondiente:

df_parte2_plataformas_medias.plot(figsize = (20, 10))

plt.title("Media de películas por año y plataforma", size = 20)
```

### Apartado 2-e: Mostrar en que plataforma está disponible una película (utilizar funciones)

```python
# Creamos el Datafrme:

df_parte2_buscar = df_parte2.loc[:, ["Title", "Netflix", "Hulu", "Prime Video", "Disney+"]]
df_parte2_buscar
```

```python
# Creamos una función con la que filtrar el Dataframe por el título que nos interesa:
# Pedimos el título en un bucle y mostramos dónde está la película

def buscar_pelicula(titulo):
    df_filtrado = df_parte2_buscar[(df_parte2_buscar["Title"] == titulo)] 
    return df_filtrado

opcion = 0

while opcion == 0:
    titulo = input('Introduzca el título de la película: ')
    loc = buscar_pelicula(titulo)
    print(loc)
    opcion = int(input('Pulse 0 buscar otro título, cualquier otro numero para terminar: '))

```

### Apartado 2-f: Muestre la media del IMDb rating por plataforma

```python
# En este último ejercicio, voy a crear varios Dataframe, uno por plataforma, calcular la media (se calculan 2 uns es
# con el IMDb que tiene con sus películas y otro es el IMDb con la de los demás). Selecciono la media de las películas 
# que están en la plataforma y lo presento en un for.

plat = ["Netflix", "Hulu", "Prime Video", "Disney+"]

for i in plat:
    df_parte2_imdb = df_parte2.loc[:, ["IMDb", i]]
    df_parte2_imdb = df_parte2_imdb.groupby(by = i).mean()
    print(f'La media IMDb de Netflix es : {df_parte2_imdb.loc[1]}')
```
